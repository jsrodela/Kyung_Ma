```python
import pygame
import sys
import random
import imageio
from PIL import Image

array = []

for i in range(1, 11):
    array.append(i)
    array.append(100-i)
array.append(200)
array.append(300)
# 초기화
pygame.init()

horse_animations = []
for i in range(1, 6):
    frames = []
    img_reader = imageio.get_reader(f"horse{i}.gif")
    
    for frame in img_reader:
        pil_image = Image.fromarray(frame)
        pil_image = pil_image.convert("RGBA")  # RGBA로 변환
        pygame_image = pygame.image.fromstring(pil_image.tobytes(), pil_image.size, pil_image.mode)
        frames.append(pygame_image)
    horse_animations.append(frames)
    
horses = [{"image" : image, "x": 50, "y": 100 + 100 * i} for i, image in enumerate(horse_animations)]

# 화면 설정
width, height = 1200, 600
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("Horse Racing Game")

# 색상 정의
black = (0, 0, 0)
white = (255, 255, 255)
red = (255, 0, 0)

# 게임 상태
class GameState:
    def __init__(self):
        self.current_screen = "user"
        self.user_rect_color = white
        self.horses = [{"image": image, "x": 50, "y": 50 + 110 * i} for i, image in enumerate(horse_animations)]
        self.winner = None
        self.race_started = False
        self.destination_x = width - 30  # 오른쪽 도착점 x값
        self.reset_horses()

    def switch_screen(self):
        if self.current_screen == "user":
            self.current_screen = "admin"
        else:
            self.current_screen = "user"
    
    def change_color(self):
        self.user_rect_color = red if self.user_rect_color == white else white
    
    def reset_horses(self):
        for horse in self.horses:
            horse["x"] = 50

    def race(self):
        if not self.winner:
            for horse in self.horses:
                horse["x"] += (random.choice(array)/25)  # 속도 더 늦추기
                if horse["x"] >= self.destination_x:
                    self.winner = [i+1 for i, h in enumerate(self.horses) if h == horse]
    
    def find_winner(self):
        if self.winner:
            return f"Winner: Horse {', '.join(map(str, self.winner))}"
        return ""

game_state = GameState()

# 버튼 생성
button_font = pygame.font.Font(None, 36)
button_rect = pygame.Rect(width // 2 - 75, height // 2 - 25, 150, 50)
button_text = button_font.render("Button", True, white)

# 게임 루프

running = True
start_ticks = 0
frame_indices = [0] * len(horse_animations)

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                if not game_state.race_started:
                    game_state.race_started = True
                else:
                    if game_state.current_screen == "user":
                        game_state.switch_screen()
                    else:
                        game_state.reset_horses()
                        game_state.switch_screen()
    
    screen.fill(black)
    
    
        
    if game_state.race_started:
        current_ticks = pygame.time.get_ticks() - start_ticks
        if current_ticks >= 100:
            game_state.race()
            start_ticks = pygame.time.get_ticks()
        
    if game_state.current_screen == "user":
        for i, horse in enumerate(game_state.horses):
            
            animation = horse_animations[i]
            frame_surface = animation[frame_indices[i]]
            
            scaled_frame_surface = pygame.transform.scale(frame_surface, (144, 100))
            
            horse_rect = scaled_frame_surface.get_rect(topleft=(horse["x"], horse["y"]))
            screen.blit(scaled_frame_surface, horse_rect)
            
            frame_indices[i] = (frame_indices[i] + 1) % len(animation)
            
            
            
    else:
        pygame.draw.rect(screen, white, button_rect)
        screen.blit(button_text, (button_rect.centerx - button_text.get_width() // 2, button_rect.centery - button_text.get_height() // 2))
    
    winner_text = game_state.find_winner()
    if winner_text:
        font = pygame.font.Font(None, 36)
        text = font.render(winner_text, True, white)
        screen.blit(text, (width // 2 - text.get_width() // 2, height // 2 - text.get_height() // 2))
    
    pygame.display.flip()

# 종료
pygame.quit()
sys.exit()
```
